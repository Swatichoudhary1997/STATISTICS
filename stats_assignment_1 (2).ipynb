```python
import pandas as pd
```

```python
# loading dataset

df=pd.read_csv('IRIS.csv')
df
```

```python
''' 1.Central Tendency
    Without directly calculating the mean, determine which sepal measurement (sepal_length or
    sepal_width) has a higher average by observing the dataset.'''
df.describe()
#by the output of above command it is clear that average(sepal_length) > average(sepal_width)
```

```python
# average(sepal_length) > average(sepal_width)
# way 2
if df['sepal_length'].median() > df['sepal_width'].median():
    print("sepal_length has a higher average")
else:
    print("sepal_width has a higher average")
```

```python
''' 2. Dispersion
    Compare the variability of petal measurements by checking the difference between the
    ranges of petal_length and petal_width. Which feature shows more spread?'''

#calculating range for petal length and petal width

range_petal_length = df['petal_length'].max() - df['petal_length'].min()
range_petal_width = df['petal_width'].max() - df['petal_width'].min()
print("Range of petal_length:", range_petal_length)
print("Range of petal_width:", range_petal_width)

# check for spreadness(higher the range, higher will be spreadness)

if range_petal_length > range_petal_width:
    print("petal_length has more spread")
else:
    print("petal_width has more spread")
```

```python
''' 3.Quartiles
    Identify which feature, sepal_length or sepal_width, has a higher interquartile range (IQR),
    indicating greater variability in the middle 50% of the data.'''

# calculating IQR

iqr_sepal_length = df['sepal_length'].quantile(0.75) - df['sepal_length'].quantile(0.25)
iqr_sepal_width = df['sepal_width'].quantile(0.75)-df['sepal_width'].quantile(0.25)
print("IQR of sepal_length:", iqr_sepal_length)
print("IQR of sepal_width:", iqr_sepal_width)

# comparing IQR OF BOTH FOR VARIABILITY( IQR is directly proportional to variance)

if iqr_sepal_length > iqr_sepal_width:
    print("IQR of sepal length is high,indicating greater variability in the middle 50% of the data")
else:
    print("IQR of sepal width is high,indicating greater variability in the middle 50% of the data")
```

```python
''' 4. Standard Deviation Insight
   (Based on the computed standard deviations of sepal_length and sepal_width, determine
  which attribute exhibits more variability and discuss what this indicates about the
  consistency of the two features.)'''

#calculating standard deviation for both

sd_sepal_length = df['sepal_length'].std()
sd_sepal_width = df['sepal_width'].std()
print("Standard Deviation of sepal_length:", sd_sepal_length)
print("Standard Deviation of sepal_width:", sd_sepal_width)

# comparing standard deviation for variability

if sd_sepal_length > sd_sepal_width:
    print("sepal_length has more variability")
else:
    print("sepal_width has more variability")

# discussing consistency

''' higher standard deviation = more variability = less consistency,
           lower standard deviation = less variability = more consistency'''

if sd_sepal_length > sd_sepal_width:
    print("sepal_width has more consistency")
else:
    print("sepal_length has more consistency")
```

```python
''' 5.Shape of Distribution
Without plotting, assess whether the petal_length distribution is skewed. Use a descriptive
statistic that measures skewness and infer the distribution shape.'''

# calculating skewness of petal length

skew = df['petal_length'].skew()
print('skewness of petal length =', skew)

# interpreting skewness

if skew>0:
  print('the distribution of petal length is positively skewed(right skewed)')
elif skew<0:
  print('the distribution of petal length is negatively skewed(left skewed)')
else:
  print('the distribution of petal length is approximately symmetrical')
```

```python
''' 6. Symmetry
   For which attribute, sepal_length or sepal_width, is the skewness closer to zero? What does
   this imply about the symmetry of its distribution?'''

# calculating skewness of sepal_length and sepal_width

skew_sepal_length = df['sepal_length'].skew()
skew_sepal_width = df['sepal_width'].skew()
print('skewness of sepal length =', skew_sepal_length)
print('skewness of sepal width =', skew_sepal_width)

# interpreting symmetry of distribution

if skew_sepal_length < skew_sepal_width:
    print('sepal_length is more symmetrical')
else:
    print('sepal_width is more symmetrical')
```

```python
''' 7. Coefficient of Variation
   Calculate the coefficient of variation (CV) for sepal_length and petal_length. Which feature
   has greater relative variability, and what does this tell you about the dataset?'''

# calculating cv of sepal_length and petal_length

cv_sepal_length = (df['sepal_length'].std() / df['sepal_length'].mean())*100
cv_petal_length = (df['petal_length'].std() / df['petal_length'].mean())*100
print('CV of sepal length =', cv_sepal_length)
print('CV of petal length =', cv_petal_length)

# comparing cv for variability (higher cv means greater variability while lower cv means less variability)

if cv_sepal_length > cv_petal_length:
    print('sepal_length has higher variability')
else:
    print('petal_length has higher variability')

# interpretation

# higher variability means wide range, more spreadness and less consistency

if cv_sepal_length > cv_petal_length:
    print('sepal_length has wide range, more spreadness and less consistency')
else:
    print('petal_length has wide range, more spreadness and less consistency')
```

```python
''' 8. Outlier Detection using IQR
Identify the outliers in the sepal_length feature using the IQR method (define outliers as
values that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR). How many outliers are
present?'''

# calculating iqr

iqr = df['sepal_length'].quantile(0.75) - df['sepal_length'].quantile(0.25)

# calculating lower and upper bound

lower_bound = df['sepal_length'].quantile(0.25) - (1.5 * iqr)
upper_bound = df['sepal_length'].quantile(0.75) + (1.5 * iqr)

# calculating and counting outliers

outliers = df[(df['sepal_length'] < lower_bound) | (df['sepal_length'] > upper_bound)]
print('number of outliers in sepal_length =', len(outliers))

```
